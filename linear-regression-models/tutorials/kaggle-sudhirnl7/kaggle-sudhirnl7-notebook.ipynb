## Import libraries

```python
import matplotlib.pyplot as plt  # Data visualization
import numpy as np  # Data manipulation
import pandas as pd  # Data manipulation
import seaborn as sns  # Data visualization
```

## Set up visualization configs

```python
plt.rcParams['figure.figsize'] = [8, 5]
plt.rcParams['font.size'] = 14
plt.rcParams['font.weight']= 'bold'
plt.style.use('seaborn-v0_8-whitegrid')
```

## Import dataset

```python
path = '../../datasets/'
df = pd.read_csv(f'{path}/insurance.csv')

print(f'Number of rows and columns in the dataset: {df.shape}')
print()

# Lets look into top few rows and columns in the dataset.
df.head()
```

## Plot **Charges** (target var) vs **BMI** (independent var)

```python
"""
For visualization purposes, we will fit a line using seaborn library only
for BMI as an independent variable and Charges as a dependent variable.
"""

sns.lmplot(x='bmi', y='charges', data=df, aspect=2, height=6)

plt.title('Charge vs BMI')
plt.xlabel('Boby Mass Index $(kg/m^2)$')
plt.ylabel('Insurance Charges')
```

## Exploratory data analysis

```python
df.describe()
```

## Check for missing values

```python
plt.figure(figsize=(12, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing values in the dataset')
```
